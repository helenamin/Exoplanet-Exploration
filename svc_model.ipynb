```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("data\exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
df.columns
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df[['koi_fpflag_nt', 'koi_fpflag_ss', 'koi_fpflag_co',
       'koi_fpflag_ec', 'koi_period', 'koi_time0bk', 'koi_impact',
       'koi_duration', 'koi_depth', 'koi_prad', 'koi_teq', 'koi_insol', 
       'koi_model_snr', 'koi_tce_plnt_num', 'koi_steff', 'koi_slogg', 
       'koi_srad', 'ra', 'dec', 'koi_kepmag']]

selected_features
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
target = df["koi_disposition"]
target_names = target.unique()
print(target_names)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(selected_features, target, random_state=42)
```

```python
print(X_train.shape, y_train.shape)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# from sklearn.preprocessing import LabelEncoder
# label_encoder = LabelEncoder()
# label_encoder.fit(y_train)
# encoded_y_train = label_encoder.transform(y_train)
# encoded_y_test = label_encoder.transform(y_test)
```

# Train the Model



```python
from sklearn.svm import SVC 
model_svc = SVC(kernel='linear')
model_svc
```

```python
# model_svc.fit(X_train_scaled, encoded_y_train)
model_svc.fit(X_train_scaled, y_train)
```

```python
# print(f"Training Data Score: {model_svc.score(X_train_scaled, encoded_y_train)}")
# print(f"Testing Data Score: {model2.score(X_test_scaled, encoded_y_test)}")
print(f"Training Data Score: {model_svc.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model_svc.score(X_test_scaled, y_test)}")
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
predictions = model_svc.predict(X_test_scaled) 
# print(classification_report(encoded_y_test, predictions))
print(classification_report(y_test, predictions))
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10, 50],
              'gamma': [0.0001, 0.0005, 0.001, 0.005]}
svc_grid = GridSearchCV(model_svc, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
# svc_grid.fit(X_train_scaled, encoded_y_train)
svc_grid.fit(X_train_scaled, y_train)
```

```python
print(svc_grid.best_params_)
print(svc_grid.best_score_)
```
